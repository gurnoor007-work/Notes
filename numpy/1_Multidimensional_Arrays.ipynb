#### Dimension
A single matrix has columns as _vectors_
$$
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
$$
$$
v_1 =
\begin{pmatrix}
1 \\
4 \\
7 \\
\end{pmatrix},
\qquad
v_2 =
\begin{pmatrix}
2 \\
5 \\
8 \\
\end{pmatrix},
\qquad
v_3 =
\begin{pmatrix}
3 \\
6 \\
9 \\
\end{pmatrix},
$$

Each Vector has three directions, hence a 3x3 Dimensional matrix

#### Difference between textbook Linear Algebra dimension and numpy array dimension
We define dimensions of matrices in Linear Algebra as explained above, but for _Numpy_, dimensions are defined as follows
1. Zero dimension array
```python
arr0 = np.array(1)
```
2. One dimension array
```python
arr1 = np.array([1, 2, 3])
```
3. Two dimension array
```python
arr2 = np.array([[1, 2, 3], 
                 [6, 7, 8]])
```
4. Three dimension array
```python
arr3 = np.array([
                 [
                  [1, 2, 3], 
                  [6, 7, 8][^1]
                 ], 
                 [
                  [1, 2, 3], 
                  [6, 7, 8]
                 ]
                ])
```

`Dimensions is basically the depth of the array, how many levels deep it goes`
> _Trick is to count the number of brackets at the end of everything_

````python
import numpy as np

zero_dim_array = np.array('A') # 0-dimensional array (a point)

print(zero_dim_array.ndim)
````

````python
arr1 = np.array([1, 2, 3]) # 1-dimensional array
arr2 = np.array([1, 2]) # Also 1-dimensional array
arr3 = np.array([1]) # Also 1 dimensional array

print(arr1, arr1.ndim)
print(arr2, arr2.ndim)
print(arr3, arr3.ndim)
````

.shape gives us the shape of the array.
It gives us the amount of "stuff" inside [] of each _layer_
<br/>
$$
(l, m, n) \;\Rightarrow\; (\text{outermost layer content},\; \text{next inner layer content},\; \text{next-to-next inner layer content})
$$

````python
# let's make an array of shape (2, 1, 3)
array = np.array([[[1, 2, 3]], [[5, 6, 1]]])

print(array.shape)
````

Accessing of elements is done by _multidimensional indexing_ (arr[][][]...)

````python
print(array[1][0][1]) #prints 6
````
