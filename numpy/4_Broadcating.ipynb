# Broadcasting
---
Numpy alters shapes of matrices while doing operations. Scaling the smaller shape to match the larger shape

```python
import numpy as np

arr1 = np.array([[1, 2, 3, 4]])
arr2 = np.array([[1], [2], [3], [4]])
print("arr1: ", arr1.shape)
print("arr2: ", arr2.shape)
```

### Rules for Broadcasting
$$ \begin{aligned} \textbf{where} \qquad &\texttt{arr1.shape} = (i, j),\\ &\texttt{arr2.shape} = (l, m), \end{aligned} $$ $$ \begin{aligned} \textbf{Condition:}\qquad &(i = l \land j = m)\\ &\quad\text{or}\\ &((i = 1 \lor l = 1) \land (j = 1 \lor m = 1)). \end{aligned} $$

Both arr1 and arr2 can be broadcasted because at 0th position, (1, 4) and at 1st position, (4, 1)
