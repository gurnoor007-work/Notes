# Aggregate Functions
---

````python
import numpy as np

arr = np.array([[1, 2, 3, 4, 0, -1], [10, 0, 100000, 20, 5, 10000000]])
````

#### 1. np.sum()

````python
print(np.sum(arr)) # gives sum of all elements regardless of dimensions
````

#### 2. np.mean()

````python
print(np.mean(arr)) # gives mean of all elements
````

### 3. np.std() - Standard Deviation

````python
print(np.std(arr))
````

#### 4. np.var() - Variance

````python
print(np.var(arr))
````

#### Others

````python
print(np.min(arr))
print(np.argmin(arr)) # Gives the position of min element
````

## Concept of Axis
---
1. axis = 0 (_Only along columns_)
```python
    np.aggregate_func(arr, axis=0)
```
2. axis = 1 (_Only along rows_)
```python
    np.aggregate_func(arr, axis=1)
```

````python
print(np.sum(arr, axis=0))
print(np.sum(arr, axis=1))
````
