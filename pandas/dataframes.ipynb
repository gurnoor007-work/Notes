### We will use dictionaries, whose values are arrays

```python
import pandas as pd
```

```python
data = {"Name": ["gurnoor", "prabhnoor", "mumma", "papa"], "Age": [19, 17, 45, 48]}
df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data, index=["person 1", "person 2", "person 3", "person 4"])
print(df)
```

- Loc properties

```python
print(df.loc["person 1"])
```

##### Add New Columns!!

```python
# new column will be gender
df["Gender"] = ["male", "male", "female", "male"]
print(df)
```

##### Add New Rows!!

```python
new_row = pd.DataFrame(
    [{"Name": "Sandy", "Age": 123, "Gender": "gay"}], index=["person 5"]
)  # Also add index for that row
df = pd.concat([df, new_row])
print(df)
```

- Add multiple rows

```python
new_rows = pd.DataFrame(
    [
        {"Name": "Sandy", "Age": 123, "Gender": "gay"},
        {"Name": "Chittad", "Age": 1, "Gender": "non-binary"},
    ],
    index=["person 6", "person 7"],
)
df = pd.concat([df, new_rows])
print(df)
```
