# Filtering of DataFrames

````python
import pandas as pd
````

````python
df = pd.read_csv("ai_student_impact.csv")
print(df)
````

#### In order to filter data, the following format is used:
```python
df_filtered = df[df["Column_Name"] > 10] # a random condition
df_filtered = df["random_string" in df["Column_name"]] # another random condition
```

````python
df_filtered = df[df["Student_ID"] == 149996]
print(df_filtered.to_string()) # must print a single row as a single row has student id like this
````

- You can also use multiple conditions
> and, or, nor, not (but not in python format) &, |, basically C-style operators

````python
df_filtered = df[(df["Student_ID"] > 149977) & (df["Major_Category"] == "Medical")]
print(df_filtered.to_string())
````
