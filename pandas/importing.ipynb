**Make a column serve as the index (_or label_)**

```python
df = pd.read_csv("ai_student_impact.csv", index_col="Student_ID")
print(df)
```

```python
# Now we can access row by using student id
student_149996 = df.loc[149996]
print(student_149996)
```

Now, if we don't want to show all the columns for a particular row, we can do it in the following way

```python
student_149996_info = df.loc[149996, ["Major_Category", "Prompt_Engineering_Skill", "Anxiety_Level_During_Exams"]]
print(student_149996_info)
```

##### We can also show a particular segment of data using loc

```python
students_A = df.loc[100004:100013, ["Major_Category", "Prompt_Engineering_Skill", "Anxiety_Level_During_Exams"]] # both 100004 and 100013 inlcuded
print(students_A)
```

`iloc also works the same, only index is used instead of labels. And we can also use steps, like 0:10:2`
