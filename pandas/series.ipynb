```python
import pandas as pd
```

*series*: Basically a column, containing various entries.

- Define a series

```python
data = [100, 400, 99]
series = pd.Series(data)
print(series)
```

- Add custom index

```python
series = pd.Series(data, index=['a', 'b', 'c'])
print(series)
```

- Select an entry by it's label (*select by index code*)

```python
print(series.loc[1]) #will not work as now the indices are set custom
```

```python
print(series.loc["a"]) #will work
```

```python
# You can also modify data using .loc
series.loc["b"] = 7000
print(series)
```

- Locate by integer position (***.iloc[]***)

```python
print(series.iloc[0]) # prints 100
```

- Filtering values

```python
data = [100, 102, 400, 500, 145, 660, 320]
series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])
print(series[series >= 200]) # Prints all entries that are greater than 200
```

- Using dictionaries

```python
calories = {"monday": 2980, "tuesday": 3100, "wednesday": 2899}
calorie_data = pd.Series(calories)
print(calorie_data) # keys act as labels
```

```python
# loc properties work the same
calorie_data.loc["monday"] += 100
print(calorie_data)
```

```python
print(calorie_data["monday"])
```
